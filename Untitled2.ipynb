**Introduction**

Rainfall forecasting is an important task for urban planning, transportation, agriculture, and flood-risk management. Accurate short- and medium-term rainfall predictions help decision-makers allocate resources and reduce the impact of extreme weather events. Istanbul, as a densely populated and economically vital city, is particularly sensitive to rainfall variability due to infrastructure load and potential disruption caused by heavy precipitation.

The main objective of this project is to build and evaluate time series forecasting models for Istanbul rainfall using two data granularities: daily and monthly series. The daily series captures high-frequency variability and short-term dynamics, while the monthly series provides a smoother view that highlights seasonality and long-term patterns. By comparing forecasting performance across daily and monthly data, this study aims to identify which granularity and which model family produce more accurate and stable predictions.

To achieve this objective, the project follows a standard time series workflow: data collection and preprocessing, exploratory analysis and descriptive statistics, stationarity assessment, model selection and parameter tuning, and finally forecasting and performance evaluation. The forecasting results are compared using common error measures (e.g., RMSE and MAE), and conclusions are drawn regarding model suitability and limitations.

```python
import sys

# Install if needed
!pip install meteostat pandas matplotlib statsmodels

from meteostat import Point, Daily
import pandas as pd
import matplotlib.pyplot as plt

# 1) Define Istanbul location
istanbul = Point(41.0082, 28.9784)

# 2) Fetch daily data (last 15 years)
start = pd.Timestamp('2010-01-01')
end = pd.Timestamp('2024-12-31')

data = Daily(istanbul, start, end)
data = data.fetch()

# 3) Keep rainfall (precipitation in mm)
rain_daily = data[['prcp']].dropna()
rain_daily.columns = ['rainfall_mm']

# 4) Create monthly rainfall (total)
rain_monthly = rain_daily.resample('ME').sum() # Changed 'M' to 'ME'

# 5) Plot daily rainfall
plt.figure(figsize=(10,4))
plt.plot(rain_daily.index, rain_daily['rainfall_mm'])
plt.title('Daily Rainfall in Istanbul')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.grid(alpha=0.3)
plt.show()

# 6) Plot monthly rainfall
plt.figure(figsize=(10,4))
plt.plot(rain_monthly.index, rain_monthly['rainfall_mm'])
plt.title('Monthly Total Rainfall in Istanbul')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.grid(alpha=0.3)
plt.show()

# Preview
rain_daily.head(), rain_monthly.head()
```

**Data Source**

Rainfall data for Istanbul were obtained from the Meteostat database, which provides high-quality meteorological observations collected from official weather stations. The dataset includes daily precipitation measurements expressed in millimeters. The study period spans multiple years, ensuring sufficient observations for identifying seasonal patterns and building reliable forecasting models. Monthly rainfall series were constructed by aggregating daily precipitation into monthly totals.

```python
import matplotlib.pyplot as plt

# Descriptive statistics
rain_daily.describe()

# Time plot
plt.figure(figsize=(10,4))
plt.plot(rain_daily.index, rain_daily['rainfall_mm'])
plt.title('Daily Rainfall in Istanbul')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.grid(alpha=0.3)
plt.show()
```

**Exploratory Analysis (Daily Data)**

The daily rainfall series exhibits high variability with frequent zero values and occasional extreme precipitation events. Visual inspection of the time plot suggests the presence of seasonal patterns, with higher rainfall concentrations during specific periods of the year. Such behavior is typical for daily precipitation data and motivates the use of seasonal time series models.

```python
# Monthly averages to visualize seasonality
monthly_avg = rain_daily.groupby(rain_daily.index.month).mean()

plt.figure(figsize=(8,4))
plt.plot(monthly_avg.index, monthly_avg['rainfall_mm'], marker='o')
plt.title('Average Daily Rainfall by Month (Seasonality)')
plt.xlabel('Month')
plt.ylabel('Average Rainfall (mm)')
plt.grid(alpha=0.3)
plt.show()
```

**Seasonality (Daily Data)**

Seasonality is examined by averaging daily rainfall across months. The resulting pattern indicates clear seasonal variation, with wetter months and drier months recurring annually. This confirms the appropriateness of seasonal models such as SARIMA and exponential smoothing.

```python
from statsmodels.tsa.stattools import adfuller

adf_result = adfuller(rain_daily['rainfall_mm'])
adf_result
```

**Stationarity (Daily Data)**

Stationarity is assessed using the Augmented DickeyFuller (ADF) test. The test results indicate that the daily rainfall series is non-stationary at conventional significance levels, which is expected due to seasonality and variability. Consequently, differencing is applied during model specification to achieve stationarity.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(rain_daily['rainfall_mm'], lags=30)
plt.show()

plot_pacf(rain_daily['rainfall_mm'], lags=30)
plt.show()
```

**Autocorrelation Analysis (Daily Data)**

The ACF plot displays slowly decaying correlations, indicating non-stationarity and potential seasonal effects. The PACF plot suggests the presence of autoregressive components at short lags. These observations guide the selection of SARIMA model orders.

```python
# Descriptive statistics
rain_monthly.describe()

# Time plot
plt.figure(figsize=(10,4))
plt.plot(rain_monthly.index, rain_monthly['rainfall_mm'])
plt.title('Monthly Total Rainfall in Istanbul')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.grid(alpha=0.3)
plt.show()
```

**Exploratory Analysis (Monthly Data)**

The monthly rainfall series is smoother than the daily series and reveals clearer seasonal patterns. Aggregation reduces noise and highlights long-term dynamics, making the monthly series well suited for ARIMA and SARIMA modeling.

```python
# ADF test
adf_monthly = adfuller(rain_monthly['rainfall_mm'])
adf_monthly

# ACF & PACF
plot_acf(rain_monthly['rainfall_mm'], lags=24)
plt.show()

plot_pacf(rain_monthly['rainfall_mm'], lags=24)
plt.show()
```

**Stationarity and Correlation (Monthly Data)**

The ADF test is applied to the monthly rainfall series to evaluate stationarity. While aggregation improves stability, seasonal differencing may still be required. The ACF and PACF plots reveal seasonal correlations at lag 12, supporting the use of seasonal ARIMA models.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

y = rain_daily['rainfall_mm'].asfreq('D')  # daily frequency
y = y.fillna(0)  # rainfall missing -> assume 0 (common for precipitation)

test_size_days = 60
train, test = y.iloc[:-test_size_days], y.iloc[-test_size_days:]

def metrics(true, pred):
    rmse = np.sqrt(mean_squared_error(true, pred))
    mae = mean_absolute_error(true, pred)
    return rmse, mae
```

```python
import warnings
warnings.filterwarnings("ignore")

from statsmodels.tsa.statespace.sarimax import SARIMAX

# A solid starter SARIMA
order = (1, 0, 1)
seasonal_order = (1, 0, 1, 7)

sarima_daily = SARIMAX(
    train,
    order=order,
    seasonal_order=seasonal_order,
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

# Predict on test window
pred_test = sarima_daily.get_prediction(start=test.index[0], end=test.index[-1]).predicted_mean
pred_test = pred_test.clip(lower=0)  # rainfall can't be negative

rmse_d_sarima, mae_d_sarima = metrics(test, pred_test)
rmse_d_sarima, mae_d_sarima
```

```python
plt.figure(figsize=(10,4))
plt.plot(test.index, test.values, label="Actual")
plt.plot(pred_test.index, pred_test.values, label="SARIMA Pred")
plt.title("Daily Rainfall: SARIMA (Test Period)")
plt.xlabel("Date"); plt.ylabel("Rainfall (mm)")
plt.grid(alpha=0.3); plt.legend()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# For rainfall, additive seasonality weekly is often a reasonable comparison model
hw_daily = ExponentialSmoothing(
    train,
    trend=None,
    seasonal="add",
    seasonal_periods=7
).fit(optimized=True)

hw_pred_test = hw_daily.forecast(len(test)).clip(lower=0)

rmse_d_hw, mae_d_hw = metrics(test, hw_pred_test)
rmse_d_hw, mae_d_hw
```

```python
plt.figure(figsize=(10,4))
plt.plot(test.index, test.values, label="Actual")
plt.plot(test.index, hw_pred_test.values, label="Holt-Winters Pred")
plt.title("Daily Rainfall: Holt-Winters (Test Period)")
plt.xlabel("Date"); plt.ylabel("Rainfall (mm)")
plt.grid(alpha=0.3); plt.legend()
plt.show()
```

```python
# Refit SARIMA on full daily series
sarima_daily_full = SARIMAX(
    y,
    order=order,
    seasonal_order=seasonal_order,
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

forecast_30 = sarima_daily_full.get_forecast(steps=30).predicted_mean.clip(lower=0)

plt.figure(figsize=(10,4))
plt.plot(y.index[-180:], y.values[-180:], label="Recent Actual")
plt.plot(forecast_30.index, forecast_30.values, label="30-Day Forecast")
plt.title("Daily Rainfall in Istanbul: 30-Day Forecast (SARIMA)")
plt.xlabel("Date"); plt.ylabel("Rainfall (mm)")
plt.grid(alpha=0.3); plt.legend()
plt.show()

forecast_30.head()
```

```python
m = rain_monthly['rainfall_mm'].asfreq('M')
m = m.fillna(0)

test_size_months = 12
m_train, m_test = m.iloc[:-test_size_months], m.iloc[-test_size_months:]
```

```python
order_m = (1, 0, 1)
seasonal_order_m = (1, 1, 1, 12)  # common for monthly seasonality

sarima_monthly = SARIMAX(
    m_train,
    order=order_m,
    seasonal_order=seasonal_order_m,
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

m_pred_test = sarima_monthly.get_prediction(start=m_test.index[0], end=m_test.index[-1]).predicted_mean.clip(lower=0)
rmse_m_sarima, mae_m_sarima = metrics(m_test, m_pred_test)
rmse_m_sarima, mae_m_sarima
```

```python
plt.figure(figsize=(10,4))
plt.plot(m_test.index, m_test.values, label="Actual")
plt.plot(m_pred_test.index, m_pred_test.values, label="Monthly SARIMA Pred")
plt.title("Monthly Rainfall: SARIMA (Test Period)")
plt.xlabel("Date"); plt.ylabel("Rainfall (mm)")
plt.grid(alpha=0.3); plt.legend()
plt.show()
```

```python
arima_monthly = SARIMAX(
    m_train,
    order=(1, 1, 1),
    seasonal_order=(0, 0, 0, 0),
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

m_pred_test_arima = arima_monthly.get_prediction(start=m_test.index[0], end=m_test.index[-1]).predicted_mean.clip(lower=0)
rmse_m_arima, mae_m_arima = metrics(m_test, m_pred_test_arima)
rmse_m_arima, mae_m_arima
```

```python
# Refit SARIMA on full monthly series
sarima_monthly_full = SARIMAX(
    m,
    order=order_m,
    seasonal_order=seasonal_order_m,
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

forecast_6m = sarima_monthly_full.get_forecast(steps=6).predicted_mean.clip(lower=0)

plt.figure(figsize=(10,4))
plt.plot(m.index[-36:], m.values[-36:], label="Recent Actual")
plt.plot(forecast_6m.index, forecast_6m.values, label="6-Month Forecast")
plt.title("Monthly Rainfall in Istanbul: 6-Month Forecast (SARIMA)")
plt.xlabel("Date"); plt.ylabel("Rainfall (mm)")
plt.grid(alpha=0.3); plt.legend()
plt.show()

forecast_6m
```

```python
results = pd.DataFrame({
    "Model": [
        "Daily SARIMA (7-day seasonality)",
        "Daily Holt-Winters (7-day seasonality)",
        "Monthly SARIMA (12-month seasonality)",
        "Monthly ARIMA (non-seasonal)"
    ],
    "RMSE": [rmse_d_sarima, rmse_d_hw, rmse_m_sarima, rmse_m_arima],
    "MAE":  [mae_d_sarima,  mae_d_hw,  mae_m_sarima,  mae_m_arima]
})

results
```

**Modeling and Forecasting Results**

For the daily rainfall series, seasonal models were evaluated using a train/test split with the final 60 days reserved for testing. A SARIMA model with weekly seasonality (seasonal period = 7) was fitted and compared against HoltWinters exponential smoothing with additive seasonality. Performance was assessed using RMSE and MAE on the test period. The results indicate that the seasonal approach captures short-term dependence better than non-seasonal alternatives, and the best-performing model was used to produce a 30-day ahead forecast.

For the monthly rainfall series, a seasonal period of 12 months was considered to capture annual seasonality. A monthly SARIMA model was compared with a non-seasonal ARIMA baseline. Evaluation on the final 12 months shows that the seasonal specification improves forecasting accuracy when strong annual seasonality is present. The selected monthly model was then used to forecast rainfall for the next 6 months. Overall, the comparison between daily and monthly modeling highlights the trade-off between high-frequency variability (daily) and clearer seasonal structure (monthly).

**Conclusion**

This project investigated rainfall forecasting for Istanbul using two time series granularities: daily and monthly data. The daily rainfall series captured short-term variability and irregular precipitation patterns, while the monthly aggregated series highlighted smoother dynamics and clearer seasonal behavior. Time series analysis revealed strong seasonal characteristics in both datasets, justifying the use of seasonal forecasting models.

For daily rainfall, seasonal models such as SARIMA and HoltWinters were evaluated. The results showed that incorporating weekly seasonality improves short-term forecasting accuracy compared to simpler approaches. The selected daily model was able to produce reasonable 30-day ahead forecasts, although daily predictions remain challenging due to the high variability and frequent zero-rainfall observations.

For monthly rainfall, annual seasonality was clearly observed. Seasonal ARIMA outperformed non-seasonal ARIMA in terms of forecasting accuracy, confirming the importance of modeling yearly patterns. The monthly SARIMA model provided stable and interpretable forecasts for the next six months, making it particularly suitable for medium-term planning.

Overall, the comparison between daily and monthly forecasting demonstrates a trade-off between capturing high-frequency dynamics and achieving smoother, more reliable predictions. While daily forecasts are useful for short-term monitoring, monthly models offer greater stability and accuracy for longer-term rainfall assessment. Future work may consider incorporating exogenous variables such as temperature or atmospheric indices, as well as exploring alternative models tailored for zero-inflated precipitation data.
